```python
train = pd.read_csv("/Users/nilanjansahu/Downloads/Machine_Learning/titanic/train.csv")
test = pd.read_csv("/Users/nilanjansahu/Downloads/Machine_Learning/titanic/test.csv")
```

```python
train.head(5)
```

```python
dataset = [train,test]
```

```python
#Feature if a person has cabin or not
import re

train["Has_Cabin"] = train['Cabin'].apply(lambda x: 0 if type(x)==float else 1)
test["Has_Cabin"] = test['Cabin'].apply(lambda x: 0 if type(x)==float else 1)

for data in dataset:
    if (data["Sex"]==" ").bool:
        data["Sex"] = data["Sex"].replace(" ", np.nan)

#Filling NA
for data in dataset:
    data['Family_Size'] = data['SibSp'] + data['Parch'] + 1
    data['Embarked'] = data['Embarked'].fillna('S')
    data['Fare'] = data['Fare'].fillna(data['Fare'].median())
    data['Age'] = data['Age'].fillna(data['Age'].median())
    data['Sex'] = data['Sex'].fillna('others')
    
train['Striped_Age'] = pd.qcut(train['Age'],4)
train['Striped_Fare'] = pd.qcut(train['Fare'],4)

#func to get title from name
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)
    return " "

#create Title column
for data in dataset:
    data["Title"] = data["Name"].apply(get_title)
    data["Title"] = data["Title"].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 
                                           'Sir', 'Jonkheer', 'Dona'], 'Rare')
    data["Title"] = data["Title"].replace('Mlle', 'Miss')
    data["Title"] = data["Title"].replace('Ms', 'Miss')
    data["Title"] = data["Title"].replace("Mme", "Mrs")
    if (data["Title"]==" ").bool:
        data["Title"] = data["Title"].replace(" ", np.nan)
        data['Title'] = data['Title'].fillna("Others")

```

```python
title_mapping={'Others': 0, 'Mr': 1, 'Miss': 2, 'Mrs': 3, 'Master': 4, 'Rare': 5}
sex_maping={'female': 0, 'male': 1, 'others': 2}

#assign numbers to sex, name, Age, Emarked and Fare
for data in dataset:
    data['Sex'] = data['Sex'].map({'female': 0, 'male': 1, 'others': 2}).astype(int)
    data['Title'] = data['Title'].map(title_mapping).astype(int)
    data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2}).astype(int)
```

```python
for data in dataset: 
    if(data['Fare'] <= 7.91).bool():
        data['Fare'] = 0
        
    elif(data['Fare']>7.91 & data['Fare']<=14.454).bool():
        data['Fare'] = 1
    
    elif(data['Fare']>14.454 & data['Fare']<=31).bool():
        data['Fare'] = 2
        
    elif(data['Fare']>31).bool():
        data['Fare'] = 3
    
    data['Fare'] = data['Fare'].astype(int)
    
    if(data['Age']<=16).bool():
        data['Age'] = 0
    
    elif(data['Age']>16 & data['Age']<=32).bool():
        data['Age'] = 1
    
    elif(data['Age']>32 & data['Age']<=48).bool():
        data['Age'] = 2
    
    elif(data['Age']>48 & data['Age']<=64).bool():
        data['Age'] = 3
    
    elif(data['Age']>64).bool():
        data['Age'] = 4
```
